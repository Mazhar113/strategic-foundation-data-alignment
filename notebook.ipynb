# Strategic Foundation EDA\nThis notebook performs basic EDA and KPI calculations on the sample CSVs.

```python
import pandas as pd\norders = pd.read_csv('data/orders.csv', parse_dates=['order_date'])\ncustomers = pd.read_csv('data/customers.csv')\norders.head()
```

```python
# Revenue by month\norders['month'] = pd.to_datetime(orders['order_date']).dt.to_period('M').dt.to_timestamp()\nrev = orders.groupby('month')['order_value'].sum().reset_index()\nrev.head()
```

```python
# Top customers\ntop = orders.groupby('customer_id')['order_value'].sum().reset_index().sort_values('order_value', ascending=False).head(20)\nprint(top)
```
